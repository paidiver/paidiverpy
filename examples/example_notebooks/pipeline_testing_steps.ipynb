## Run a pipeline with test inputs

This process will run the entire pipeline described in a configuration file and we will add some  test input to try some metrics

```python
%load_ext autoreload
%autoreload 2
from paidiverpy.pipeline import Pipeline
from paidiverpy.resample_layer import ResampleLayer
```

```python
pipeline = Pipeline(config_file_path="../config_files/config_benthic2.yaml")
```

```python
# See the pipeline steps. Click in a step to see more information about it
pipeline
```

```python
# See the pipeline steps in json format. As you can see, the Overlapping step is on test mode
pipeline.steps
```

```python
# Run the pipeline. You will see a plot of the overlapping.
# This plot will help you to decide if you are ok with the limits provided in the overlapping( theta = 40, omega - 80 and overlapping of 50%)
pipeline.run()
```

```python
# See the images output.You can see that the step was not added to the images output
pipeline.images
```

```python
# Change the step to another limits: overlapping of 70%
pipeline.add_step(
    "overlapping",
    ResampleLayer,
    {
        "mode": "overlapping",
        "params": {"theta": 40, "omega": 57, "threshold": 0.7},
        "test": True,
    },
    1,
    substitute=True,
)
```

```python
pipeline.steps
```

```python
# Rerun the pipeline from the second step. It is still on test mode
pipeline.run(from_step=0)
```

```python
# I am happy with the new value. So now I can update the value on the overlapping step (test to false), and run again
pipeline.add_step(
    "overlapping",
    ResampleLayer,
    {
        "mode": "overlapping",
        "params": {"theta": 40, "omega": 57, "threshold": 0.7},
        "test": False,
    },
    1,
    substitute=True,
)
pipeline.run()
```

```python
# now the images were updated
pipeline.images
```

```python
# Now I am adding a new step to resample based on obscure images. It is in test mode
pipeline.add_step(
    "datetime",
    ResampleLayer,
    {
        "mode": "datetime",
        "params": {"min": "2018-06-11 04:14:00", "max": "2018-06-11 04:20:00"},
        "test": True,
    },
)
```

```python
pipeline
```

```python
pipeline.run()
```

```python
# I am happy with the limits. So I will apply the changes and rerun
pipeline.add_step(
    "datetime",
    ResampleLayer,
    {
        "mode": "datetime",
        "params": {"min": "2018-06-11 04:14:00", "max": "2018-06-11 04:20:00"},
        "test": False,
    },
    2,
    substitute=True,
)
pipeline.run()
```

```python
# Now the outputs were save to the images
pipeline.images
```
